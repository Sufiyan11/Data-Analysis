### I. Numpy

#### 1. Concatenate two 2d arrays.

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])

concatenated_array = np.concatenate((array1, array2), axis=1)
print(concatenated_array)

concatenated_array = np.concatenate((array1, array2), axis=0)
print(concatenated_array)

concatenated_array = np.hstack((array1, array2))
print(concatenated_array)

concatenated_array = np.vstack((array1, array2))
print(concatenated_array)
```

### III. National University Rankings.

```python
data = pd.read_csv("National_Universities_Rankings.csv")
```

#### 1. Find the universities along with state whose fee is in between 25,000 to 30,000.

```python
data["TuitionFees"] = [int("".join(filter(str.isdigit, s))) for s in data["Tuition and fees"]]

df = data[(data["TuitionFees"] >= 25000) & (data["TuitionFees"] <= 30000)]

display(df[['Name', 'Location', 'Tuition and fees']])
```

#### 2. Find university where undergraduate enrollement is more than 25000 containing in-state students.

```python
data["UndergradEnrollment"] = [int("".join(filter(str.isdigit, s))) for s in data["Undergrad Enrollment"]]

df = data[data['UndergradEnrollment'] > 25000][data['In-state'].notnull()]

display(df[['Name', 'Location', 'In-state', 'Undergrad Enrollment']])
```

#### 3. Find the states where universities are located in three or more cities.

```python
data["state"] = [i[-2:] for i in data['Location']]

city_counts = data.groupby("state")["Name"].count()

display(city_counts[city_counts >= 3].reset_index())
```

#### 4. Find max and min tuition fee in each state.

```python
display(data.groupby('state').agg({'Tuition and fees': ['max', 'min']}))
```
